# Part A: Pipetting, Accuracy, and Precision
Note: At standard temperature and pressure, 1mL H<sub>2</sub>O = 1g
### Pipette calibration protocol
For each set of measurements: 
  - Record mass of an empty 1.5mL eppendorf tube.
  - Add specified volume of water, record mass.
  - Repeat previous step 4 more times (5 total).
  - For each of the previous 5 measurements, $\Delta$ mass = (current mass - previous mass)
  - Calculate expected $\Delta$ mass, average of $\Delta$ mass, converted pipetted volume, stdev, and % error.
  - Converted volume ($\mu$L) = $\Delta$ mass (g) * 1000
  
Measurements to take:
  - Using P2 perform measurements for 0.5$\mu$L and 2$\mu$L.
  - Using P20 perform measuremnts for 2$\mu$L and 20$\mu$L.
  - Using P200 perform measuremtns for 20$\mu$L and 200$\mu$L.
  - Using P1000 perform measurements for 200$\mu$L and 1mL. NOTE: use a 15mL falcon tube for the 1000mL measurements.

**NOTE:** I have very little faith in the $<20\mu$L measurements, since I could watch the balances tick up and down more than pipetting amounts just from environmental vibrations.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

red_half=[.0008,.0006,.0005,.0006,.0005]
blue_half=[.0009,.0002,.0004,.0002,.0002]
water_half=[.0002,.0012,-.0001,.0002,.0007]
red_5=[.0062,.0065,.0058,.0049,.0066]
blue_5=[.0025,.0054,.0056,.0055,.0055]
water_5=[.0051,.0051,.0054,.0049,.0048]

def plot_dist(red,blue,water,x):
    fig=plt.figure(figsize=(16,8))
    plt.bar([1,2,3,4,5], red, color='red', label="red liquid")
    plt.bar([6,7,8,9,10], blue, color='blue', label="blue liquid")
    plt.bar([11,12,13,14,15], water, color='gray', label="water")
    plt.xlabel(x)
    plt.legend()
    plt.xlim()
    plt.ylabel("mass (g)")
    plt.show()
    #fig.savefig(file+"_plot.pdf")
    
plot_dist(red_half,blue_half,water_half,"0.5 uL trials")
plot_dist(red_5,blue_5,water_5,"5 uL trials")

```

The data is too noisy for me to draw any strong conclusions about precision, but the red liquid appears to be denser and more viscous than water, and that seems to cause wider variation in pipetted amounts. Unsurprisingly, my pipetting (and/or the balance) is more precise at larger volumes.

# Part C: Dilutions, working with the Qubit
For all measurements, I used 1$\mu$L of diluted sample and 199$\mu$L of Qubit working solution. So to get from ng/mL to ng/$\mu$L I used 0.2\*ng/mL\*(dilution factor). According to [Thermo Fisher](https://www.thermofisher.com/us/en/home/industrial/spectroscopy-elemental-isotope-analysis/molecular-spectroscopy/fluorometers/qubit/qubit-assays.html) the detection range for the Qubit RNA HS Assay Kit is 5-100ng, which would be 25-500ng/mL. 

If the first readings were accurate, I would only need a 2-fold dilution to get within the linear range. The 1:2 dilutions didn't give anything close to half the 1:1 values, so I'm pretty sure we're not in the linear range. Likewise, the 1:4 dilution values weren't half of the 1:2 values.

Thermo Fisher's site says the detection range for the Qubit RNA HS Assay Kit is 5-100ng in the 200$\mu$L sample, which would be 25-500ng/mL. This would make the 1:8 and 1:16 measurements within the linear range. However, it really didn't seem like I was getting consistent answers, so I looked at the website and found they also give the linear range as "between 250 pg/µL and 100 ng/µL" which is different than 5-100ng total in the sample. So I did three more dilutions (1:32,1:64,1:128) and that seemed to reach the linear range. It is possible that the time the samples were sitting around affected the results slightly.

The Qubit gives ng/mL, so I calculated sample concentration in ng/$\mu$L as 0.2\*(Qubit value)\*(dilution factor)

The last two dilutions (1:64,1:128) gave relatively similar concentrations, so I think the true concentration is ~1100ng/$\mu$L.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

dils=[1,2,4,8,16,32,64,128]
d1=[170,260,448,555.2,672,889.6,1024,1049.6]
d2=[132,264,432,640,768,1008.8,1088,1075.2]

def plot_dist(dil,val1,val2):
    fig=plt.figure(figsize=(16,8))
    plt.scatter(dil,val1, label="first replicate")
    plt.scatter(dil,val2, label="second replicate")
    plt.xlabel("dilution factor")
    plt.legend()
    plt.xlim()
    plt.ylabel("calculated concentration (ng/uL)")
    plt.show()
    #fig.savefig(file+"_plot.pdf")
    
plot_dist(dils,d1,d2)
```
