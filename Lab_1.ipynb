# **Part A Protocol**
### Equipment: 4 Pipets (1 mL, 200 uL, 20 uL, 2 uL) w/ appropriate size tips, Balance, Weigh Boats.   
### Materials: Deionized Water, Unknown Blue Fluid, Unknown Red Fluid  
**_Note: Wear gloves when handling equipment._**  
1. Sterilize work bench with 70% ethanol.  
2. Prepare an appropriate receptacle by filling it with more than enough deionized water to complete experiment.  
3. Place an empty weighboat on the balance, then tare the balance.  
4. Beginning with the largest pipet (1 mL):  
    a. Pipet its full volume of water from receptacle into the weighboat that is atop the balance. Record the weight in lab notebook.  
    b. Repeat steps 3 & 4a two more times, so that you have a total of 3 trial weights for the pipet's full volume.  
5. Place an empty weighboat on the balance, then tare the balance.  
6. Now, set the pipet to hold exactly 1/2 of its volume.  
    a. Pipet exactly 1/2 of the full pipet volume of water from the receptacle into the weighboat that is atop the balance. Record the weight in the lab notebook.
    b. Repeat steps 5 & 6a two more times, so that you have a total of 3 trial weights for 1/2 the pipet's full volume.
7. Repeat steps 3-6 for each size of pipet.  
_Note: 1uL was too small a quantity to measure with the balance, so these measurements were not taken._

## Qubit

_What is the concentration of your original sample?_
> 1. I first diluted 1 uL of original sample with water, to 5 uL.  
2. I then diluted 1 uL from this new solution to 2 uL. 
3. I added 198 uL of working solution to the 2 uL sample. The Qubit gave a measurement of 2.94 ug/mL.  

> So, if we double this number and multiply the result by 5, the original concentration of my sample was: (2.94 * 2)*5 = 29.4 ug/mL.    

_What is the detection range of the Qubit for this assay?_  
The detection range of the Qubit for the broad range assay is 20 ng - 1000 ng. This was found by Google search.  

Because the range is given as mass, we must consider the amount of RNA in the sample. If we know our concentration in mass/volume, we can control how much RNA mass we include in the sample. If I have an original sample with a concentration of 500 ng/uL and I remove 1 uL to combine with 199 uL working solution, then I know I will have 500 ng RNA, and that I am consequently within the range of the assay.  

My sample was out of range. I made a 5 fold dilution. I then extracted a uL from this and made a 2 fold dilution. In total, I diluted my sample by a factor of 10.  

2-fold dilutions **were** roughly 2-fold lower than the concentrated sample. However, the 3rd round of 2-fold dilutions was out of range, being too low for Qubit. This was not according to calculations made in advance, so some error is likely to have taken place at this step. Possible errors include, human error in pipet measurement, RNA failing to mix thoroughly with working solution, or RNA degradation.


```python
plt.figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
dil_factor = [5,10,10]
conc = [5.0,2.79,2.84]
plt.plot(dil_factor,conc,'r^')
plt.xlabel("Dilution Factor (1:n)")
plt.ylabel("Measured Concentration (ug/mL)")
plt.title("Relationship Between Dilution Factor and Measured Concentration")
plt.xlim([4,11])
plt.ylim([2,6])
```
