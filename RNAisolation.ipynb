### Graph the Absorbance Spectrum Using Matplotlib    
---  

```python
% matplotlib inline
import matplotlib.pyplot as plt
file = "2018_07_26_RNA_group2-3_spectra.tsv"
x = []
y = []
with open(file) as f:
    LC = 0
    for line in f:
        LC += 1
        line = line.strip()
        if LC >= 4:
            wavelen, absorb = line.split('\t')
            x.append(float(wavelen))
            y.append(float(absorb))

peak = max(y)
xpos = y.index(peak)
peakx = x[xpos]
xtick = []
for n in range(220,351,5):
    xtick.append(n)
ytick = []
for n in range(0,71,2):
    ytick.append(n)
plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x,y,'r')
plt.plot(peakx,peak,'bD')
plt.grid(True)
plt.title('Absorbance Spectrum')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.xticks(xtick,xtick);
plt.yticks(ytick,ytick);
```

_How does a TRIzol extraction work –i.e. what is the purpose of each step and how does it achieve the desired results?_    
    
>In the first step, TRIzol is added to the RNA pellet, in order to dissolve the RNA pellet and denature the proteins. Next, chloroform is added to cause phase separation. Chloroform achieves this separation because it is miscible in phenol, but has a higher density. So, after centrifuging, the sample separates into 3 phases. RNA remains on top, in the aqueous phase. Beneath this is the interface, where DNA is found, and at the bottom is the organic phase, where proteins are found. This separation is caused by differences in the density between proteins, DNA, and RNA, in concert with the differences in density of phenol and chloroform. Next, remove the aqueous phase containing the RNA via pipet. Then, isopropanol is added into the recently removed aqueous phase. Because RNA is insoluble in isopropanol, it aggregates. Upon centrifugation, an RNA pellet is formed at the bottom of the tube. Next, this pellet is washed with a 70% ethanol solution. This ethanol helps to remove impurities from the sample. It's method of action is similar to isopropanol. Because RNA is insoluble in ethanol, the RNA aggregates. After centrifugation, the RNA pellet remains on the bottom of the tube. The ethanol wash is then discarded, and the sample is allowed to air-dry. This is important because ethanol is another sample contaminant. Allowing all ethanol to evaporate removes this contaminant from your sample. Finally, the RNA is resuspended in RNase-free water, to be stored at -20 degrees Celsius.

The Qubit only measures the fluorescence caused by reagent binding to RNA, and also cannot tell us about DNA contamination. However, if we want to ensure DNA is not in our sample, we could use ThermoFisher's TURBO DNA-free Kit. DNA removal is accomplished by DNases, which cut up the DNA into ever smaller bits.

_How might the methods directly above affect your sequencing results?_  
        
> The presence of RNases would degrade the RNA in your sample, by cleaving it. This would result in shorter reads and the loss of some data contained in the lost longer reads. _What is the difference between using the absorbance at 260 nm and the Qubit (i.e. how do they work, what are they detecting)? Which method is best? Why?_  

>Absorbance at 260 nm measures the absorbance of RNA and DNA, which both absorb at this wavelength. It is useful for its 260/230 and 260/280 ratios, which tell the researcher if there are impurities in the sample. However, it does not tell the researcher if the RNA is degraded, because degraded RNA also absorbs at 260nm. The Qubit, on the other hand, measures fluorescence instead of absorbance. The proprietary Qubit reagent fluoresces when it binds to RNA. The Qubit measures the intensity of this fluorescence to determine RNA concentration. Of the two, the Qubit is much more accurate in its reporting of concentration. But, not much direct information about contaminants or impurities is gathered. In this case, neither method is superior, since they are used for different purposes. Absorbance is good for detecting foreign particulate in your sample. Fluorescence is good for determining the concentration and amount of RNA in your sample.

_Why would you run a denaturing gel electrophoresis instead of native?_    

> RNA has a tendency to fold on itself, which would make fragments travel slower through the gel. Using a denaturing gel helps the RNA remain linear and single-stranded, so that it can travel ideally through the gel matrix.  

_How can a spectrophotometer be used to obtain your RNA’s concentration? What is the mathematical relationship/equation?_  

> **C = A/(E\*l)**, where C is the concentration of nucleic acid, A is the absorbance at a specific wavelength, E is the extinction coefficient, which for RNA is 0.025(mg/mL)^(-1)(cm)^(-1). For our sample: **C = 67.75/(1\*0.025) = 2710 ng/uL**.
